```python
# Program: Reciept calculator
# Author: Gurtissimo Gurt
# Purpose: To calculate total, tax, tip and a grand total for a meal.

priceText = input("Meal price: $")
price = float(priceText)

taxRate = 0.02
tipRate = 0.18

tax = price * taxRate
tip = price * tipRate

grandTotal = price + tax + tip

print("-------------------------------------")
print(f"Your subtotal is: ${price:.2f}.")
print(f"Your tax (2%) is: ${tax:.2f}.")
print(f"Your tip (18%) is: ${tip:.2f}.")
print(f"Your grand total is: ${grandTotal:.2f}.")
print("-------------------------------------")
```
