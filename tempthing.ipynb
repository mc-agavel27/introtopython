```python
# Program: Temperature check
# Purpose: ask user for temperature, and store a few booleans describing it.

temp = float(input("What is the temperature in fahrenheit?"))

isFreezing = temp <= 32
isCold = temp < 50
isCool = temp <= 51
isWarm = temp > 60
isHot = temp > 75

print(f"Temperature: {temp}F")
print(f"Freezing? {isFreezing}.")
print(f"Cold? {isCold}.")
print(f"Cool? {isCool}.")
print(f"Warm? {isWarm}.")
print(f"Hot? {isHot}.")
```
