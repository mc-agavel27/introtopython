## Section B — Storing comparisons in boolean variables

### Problem 3 — Voting age check

Ask the user for their `age` (convert to `int`). Then store **two** booleans:

- `isAdult` — true if age is 18 or over
- `isTeen` — true if age is between 13 and 19, inclusive (use a chained comparison: `13 <= age <= 19`)

Print both with f-strings. Example:

```
Age: 17
isAdult: False
isTeen: True
```

````python
# Your code here
age = int(input("Age:"))

isAdult = age >= 18
isTeen = 13 <= age <= 19

print(f"isAdult: {isAdult}")
print(f"isTeen: {isTeen}")
````

### Problem 4 — Temperature flags

Ask the user for a temperature in **Fahrenheit** (convert to `float`). Build four boolean flags:

- `isFreezing` — temperature is at or below 32
- `isCold` — temperature is below 50
- `isWarm` — temperature is at or above 70
- `isHot` — temperature is at or above 90

Print all four. Example:

```

  isFreezing: False
```

````python
# Your code here
temp = float(input("Temperature:"))

isFreezing = temp <= 32
isCold = temp < 50
isWarm = temp >= 70
isHot = temp >= 90

print(f"isFreezing: {isFreezing}")
print(f"isCold: {isCold}")
print(f"isWarm: {isWarm}")
print(f"isHot: {isHot}")
````

## Section D — Mixing arithmetic with comparisons

### Problem 7 — Even or odd

Ask the user for an integer. Use the `%` operator to get the remainder when divided by 2. A number is even when that remainder equals 0.

Store `isEven` as a boolean and print:

```
Enter a whole number: 14
14 % 2 = 0
isEven: True
```

Hint: `isEven = number % 2 == 0`

````python
# Your code here
num = int(input("Enter a whole number:"))

print(f"{num} % 2 = {num % 2}")
isEven = (num % 2) == 0
print(f"isEven: {isEven}")
````

### Problem 8 — Grade flags

Ask the user for their `grade` as a percentage (float, like `87.5`). Build three booleans:

- `isPassing` — grade is 60 or higher
- `isHonorRoll` — grade is 90 or higher
- `isPerfect` — grade equals 100

Print everything with the grade shown to **1 decimal place**:

```
Grade: 87.5%
isPassing:    True
isHonorRoll:  False
isPerfect:    False
```

````python
# Your code here
grade = float(input("Enter grade:"))

isPassing = grade >= 60
isHonorRoll = grade >= 90
isPerfect = grade == 100

print(f"Grade: {grade:.2f}%")
print(f"isPassing: {isPassing}")
print(f"isHonorRoll: {isHonorRoll}")
print(f"isPerfect: {isPerfect}")
````
