Rule to remember: **`input()` always returns a string.** If you need a number, wrap it in `int()` or `float()`.

## Exercise 1 — Greeting with input

Ask the user for their first name, then print a greeting that includes their name. Example output:

```
What is your first name? Alex
Hello, Alex — welcome to Python!
```

````python
name = str(input("What is your first name?"))

print(f"Hello, {name} - welcome to Python!")
````

## Exercise 3 — Age in 10 years

Ask the user their age, **convert it to an int**, add 10, and print the result. Example:

```
How old are you? 17
In 10 years you'll be 27.
```

````python
number = int(input("How old are you?"))

print(f"In 10 years you'll be {number + 10}.")
````

## Exercise 4 — Tip calculator

Ask the user for the price of a meal (use `float()` — could be something like `12.50`). Multiply by `0.20` to get a 20% tip. Print both the tip amount and the total.

Example:

```
Price of meal: 20.00
Tip (20%): 4.0
Total: 24.0
```

````python
price = float(input("Price of meal:"))
tip = 1.2
print(f"Tip: (20%): {price * (tip - 1)}")
print(f"Total: {price * tip}")
````

## Exercise 5 — Age-this-year calculator

Use `currentYear = 2026`. Ask the user for the year they were born, convert it to `int`, subtract to get their age, and print a message using the age. You must use `str()` at least once to concatenate the age into a string.

Example output:

```
What year were you born? 2008

```

````python
currentYear = 2026

year = int(input("What year were you born?"))

print(f"Your predicted age is {currentYear - year}.")
````
