### Problem 4 — Celsius to Fahrenheit

$$F = C \times \frac{9}{5} + 32$$

```

```

Use an f-string for the output.

````python
# Your code here
currentYear = 2026
gradYear = int(input("What year do you graduate?"))

if gradYear > currentYear:
    print(f"Only {gradYear - currentYear} year(s) to go!")
if gradYear == currentYear:
    print(f"This year you graduate!")
if gradYear < currentYear:
    print(f"You've already graduated!")
````

### Problem 9 — Discount calculator

Ask the user for an `originalPrice` (float) and a `discountPercent` (float, like `25` for 25%). Calculate:

- the dollar amount saved
- the final sale price

Print both using an f-string with **2 decimal places**. Example:

### Problem 10 — BMI calculator

Ask the user for their weight in **pounds** (float) and height in **inches** (float). Calculate BMI using:

$$BMI = \frac{weight}{height^2} \times 703$$

Print the result rounded to **1 decimal place**. Example:

```
Weight (lbs): 150
Height (in): 68
Your BMI is 22.8
```

Hint: use `**` for `height^2`.

### Problem 11 — Bill splitter

Ask the user for three things:

- `mealTotal` (float) — the total cost of the meal
- `tipPercent` (float) — like `18` for 18%
- `numPeople` (int) — how many people are splitting it

Calculate the tip amount, the grand total, and what each person owes. Print all three with **2 decimal places**. Example:

```
Meal total: 60
Tip percent: 20
Number of people: 4
Tip: $12.00
Grand total: $72.00
Per person: $18.00
```

````python
# Your code here
mealTotal = float(input("Meal total:"))
tipPercent = float(input("Tip percent:"))
numPeople = int(input("Number of people:"))

tip = mealTotal * (tipPercent / 100)

print(f"Tip: ${tip:.2f}")
print(f"Grand total: ${(tip + mealTotal):.2f}")
print(f"Per person: ${((tip + mealTotal) / numPeople):.2f}")
````

### Problem 15 — Mini receipt generator (capstone)

Write a small program that asks the user for:

- `itemName` (str)
- `quantity` (int)
- `unitPrice` (float)
- `taxPercent` (float) — like `7.25` for 7.25%

Then calculate:

- `subtotal = quantity * unitPrice`
- `tax = subtotal * (taxPercent / 100)`
- `total = subtotal + tax`

Print a clean receipt using f-strings, with all dollar amounts to **2 decimal places**. Example:

```
------------------------------
         RECEIPT
------------------------------
Item: Coffee
Quantity: 3
Unit price: $4.50
Subtotal: $13.50
Tax (7.25%): $0.98
Total: $14.48
------------------------------
```

Add a header comment block at the top of your code with `Program:`, `Author:`, `Date:`, and `Purpose:` (just like Lesson 01).

````python
# Your code here
itemName = input("Item Name:")
quantity = int(input("Quantity:"))
unitPrice = float(input("Unit price:"))
taxPercent = float(input("Tax percent:"))

subTotal = unitPrice * quantity
tax = (subTotal * (taxPercent / 100))
total = subTotal + tax

print("------------------------------")
print("           RECEIPT            ")
print("------------------------------")
print(f"Item: {itemName}")
print(f"Quantity: {quantity}")
print(f"Unit price: ${unitPrice:.2f}")
print(f"Subtotal: ${subTotal:.2f}")
print(f"Tax ({taxPercent}%): ${tax:.2f}")
print(f"Total: ${total:.2f}")
print("------------------------------")
````
